# Data Understanding

Data Understanding merupakan tahap kedua dalam metode CRISP-DM setelah Business Understanding. Pada tahap ini, fokus utama adalah mengenali dan memahami data yang akan digunakan dalam proyek data mining.

Sebelum melakukan pengolahan atau pembuatan model, praktisi data perlu mengetahui isi, struktur, serta kualitas data yang tersedia. Pemahaman ini penting agar proses analisis selanjutnya dapat berjalan dengan tepat dan sesuai tujuan.

Tahap Data Understanding biasanya terdiri dari beberapa kegiatan utama sebagai berikut:

---

## 1. Pengumpulan Data (Collect Initial Data)

Langkah pertama adalah mengumpulkan data yang relevan dengan kebutuhan proyek. Data dapat berasal dari berbagai sumber, seperti database perusahaan, file laporan, sistem transaksi, atau sumber eksternal lainnya.

Beberapa hal yang diperhatikan pada tahap ini:

- Sumber data yang digunakan  
- Jumlah data yang tersedia  
- Format penyimpanan data  
- Kesesuaian data dengan tujuan proyek  

Pengumpulan data harus dilakukan dengan memastikan bahwa data tersebut memang mendukung analisis yang akan dilakukan.

---

## 2. Mendeskripsikan Data (Describe Data)

Setelah data diperoleh, langkah berikutnya adalah mempelajari karakteristiknya. Pada tahap ini dilakukan pengamatan terhadap struktur dan isi data.

Beberapa hal yang biasanya dianalisis:

- Jumlah baris dan kolom  
- Nama dan jenis atribut  
- Tipe data (angka, teks, tanggal, dan sebagainya)  
- Gambaran umum isi data  

Tujuan dari proses ini adalah untuk memahami bentuk data secara menyeluruh sebelum masuk ke tahap pembersihan.

---

## 3. Eksplorasi Data (Explore Data)

Eksplorasi data dilakukan untuk menemukan pola awal, hubungan antar variabel, atau kemungkinan adanya kejanggalan dalam data.

Kegiatan yang dilakukan antara lain:

- Melihat distribusi nilai dalam setiap atribut  
- Mengamati hubungan antar variabel  
- Mengidentifikasi pola tertentu  
- Mendeteksi nilai yang tidak wajar  

Melalui eksplorasi ini, praktisi data dapat memperoleh gambaran awal mengenai potensi insight yang bisa dihasilkan.

---

## 4. Verifikasi Kualitas Data (Verify Data Quality)

Tahap ini bertujuan untuk mengecek apakah data memiliki masalah yang dapat memengaruhi hasil analisis.

Beberapa permasalahan yang umum ditemukan:

- Data kosong (missing value)  
- Data ganda (duplikasi)  
- Nilai yang tidak konsisten  
- Kesalahan pencatatan  
- Data yang tidak relevan  

Jika ditemukan masalah, maka akan menjadi bahan pertimbangan untuk tahap Data Preparation.

---

## Kesimpulan

Data Understanding adalah tahap yang berfokus pada pemahaman isi dan kualitas data sebelum dilakukan proses pengolahan lebih lanjut.

Tahap ini membantu memastikan bahwa data yang digunakan benar-benar sesuai dengan kebutuhan proyek. Dengan memahami data secara menyeluruh, proses analisis dan pembuatan model dapat dilakukan dengan lebih efektif dan akurat.

### i. Outlier

```python
print("\nDeteksi Outlier (IQR Method):")

for col in numeric_cols.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    
    print(f"{col} = Jumlah Outlier:", len(outliers))
```

- Q1 dan Q3 : Nilai kuartil 1 (25%) dan kuartil 3 (75%) tiap kolom numerik.
- IQR = Q3 - Q1 : Jarak interkuartil, ukuran sebaran tengah data.
- lower dan upper : Batas bawah dan atas untuk mendeteksi outlier:
  - Bawah: Q1 - 1.5*IQR
  - Atas: Q3 + 1.5*IQR
- outliers = df[(df[col] < lower) | (df[col] > upper)] : Memilih baris yang nilainya di luar batas.
- len(outliers) : Menghitung jumlah baris yang termasuk outlier.

## 7. Menghitung Jarak Data IRIS

### a. Menggunakan Python

```python
import numpy as np

# Ekstraksi fitur numerik dari dataset
feature_matrix = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].values

sample_A = feature_matrix[0]
sample_B = feature_matrix[1]

euclid_dist = np.linalg.norm(sample_A - sample_B)

manhattan_dist = np.sum(np.abs(sample_A - sample_B))

print("Perbandingan jarak antar dua sampel data:")
print("==========================================")
print(f"√ Jarak Euclidean  : {euclid_dist:.4f}")
print(f"√ Jarak Manhattan  : {manhattan_dist:.4f}")
print("==========================================")
```

- feature_matrix : Mengambil semua kolom numerik sebagai array NumPy untuk perhitungan jarak.
- sample_A & sample_B : Mengambil dua baris pertama sebagai sampel untuk dibandingkan.
- np.linalg.norm(sample_A - sample_B) : Menghitung jarak Euclidean antar dua sampel.
- np.sum(np.abs(sample_A - sample_B)) : Menghitung jarak Manhattan antar dua sampel.
